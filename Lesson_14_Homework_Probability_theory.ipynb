```python
# Комбинаторика.

# Задача 1.
k = np.array(3)
n = np.array(30)
print(f"1. Размещение, кол-во вариантов выбора старосты, замстаросты и профорга: {perm(n, k, exact=True)}")

# Задача 2.
answer = 2**10
print(f"2. Возведение в степень, вариантов распределения писем: {answer}")
'''
Просто ради интереса усложним задачу. При тех же вводных, каждый из почтальонов может взять не менее 3 писем.
Значит мы убираем все варианты с 0-2 письмами для каждого почтальона и количество вариантов в таком случае будет:
2 почтальона ** 10 писем - 2 почтальона*(comb(10,0) + comb(10,1) + comb(10,2)) = 1024 - 2*(1+10+45) = 912 вариантов
По желанию, тут можно поиграться и поменять количество почтальонов, количество писем и минимальное количество писем в одни руки.
'''

# Задача 3.
'''
Единственное обязательное условие - достать 1 деталь нужного сорта, остальные детали можно доставать как угодно, сколько угодно раз. К примеру,
мы можем как достать только 1 деталь 1 сорта, так и достать её в группе деталей 22321, так и даже продолжить доставать дальше детали 1322232...23.
Задача сводится к подсчёту количества всех вариантов, при которых:
а) как максимум из 50+20+1 попыток деталей достать одну 1 сорта,
б) как максимум из 30+20+1 попыток достать одну деталь 2 сорта.
Одна нужная деталь всегда будет относиться к любой комбинации, поэтому нужно посчитать сумму всех вариантов доставания деталей других сортов.
'''
answer = ((2**50)*(2**20))+((2**30)*(2**20))
print(f"3. Сумма произведений комбинаций для 1-сортной и 2-сортной детали : {answer}")

# Задача 4. Отсутствует
print(f"4. {None}")

# Задача 5.
n = np.array(7)
answer = factorial(n, exact=True)
print(f"5. Факториал, количество перестановок: {answer}")

# Задача 6.
k = np.array(5)
n = np.array(10)
print(f"6. Возведение в степень, кол-во вариантов распределения призов: {n**k}")

# Задача 7.
k = np.array(2)
n = np.array(16)
print(f"7. Сочетание, количество партий : {comb(n, k, exact=True)}")

# Задача 8.
k = np.array(5)
n = np.array(10)
print(f"8. Сочетание, количество вариантов для вручения одинаковых призов фильмам: {comb(n, k, exact=True)}")

# Задача 9.
'''
Каждый день нужно сажать минимум 1 дерево, соответственно нам нужно рассчитать количество комбинаций для остальных 3 деревьев.
Задача решается подсчётом количества разбиений, методом "stars and bars". Для логики, манульно подготовим все возможные разбиения n = 3 деревьев на k = 3 дня:
1) 0 111 0
2) 0 11 1
3) 0 1 11
4) 0 0 111
5) 1 0 11
6) 1 1 1
7) 1 11 0
8) 11 1 0
9) 11 0 1
10) 111 0 0
Итого, получается 10 разбиений. В формульном виде, способ подразумевает использование сочетания по общей формуле:
С(n + k - 1;k - 1) = C(3 + 3 - 1;3 - 1) = C(5;2), где 5 - количество "шаров", 2 - количество "перегородок".
'''
answer = comb(5, 2, exact=True)
print(f"9. 'Шары и перегородки', количество сочетаний посадки деревьев при непустом дне: {answer}")

# Задача 10.
'''
Аналогично предыдущей задаче. Есть четырёхзначное число, т.е. 4 "корзины", также есть условие, что сумма цифр из числа =5, соответственно
мы применяем сочетание разбиений 5 по 4 элементам:
0 1 0 1111
1 0 11 11
11 1 0 11 и т.д.
Обязательного условия содержания как минимум одной единицы на любой позиции нет, поэтому вычисления констант для элементов не требуется:
С(5 + 4 - 1; 4 - 1) = С(8;3)
'''
answer = comb(8, 3, exact=True)
print(f"10. 'Шары и перегородки', количество сочетаний при возможных пустых элементах {answer}")

# Задача 11.
'''
Заданы чёткие условия по количеству людей в группах. Логика такая: мы выдёргиваем 6 людей из 25, затем выдёргиваем 9 людей из оставшихся 19,
а остальные 10 будут остаточными для каждой группы разбиений. Для формирования группы А необходимо количество разбиений 25 человек по 6 элементам.
Затем каждому такому разбиению будут кратно пренадлежать разбиения группы Б по 9 элементам из оставшихся 19 человек.
'''
answer = comb(25, 6, exact=True) * comb(19, 9, exact=True)
print(f"11. Комплексное сочетание, количество вариантов разбиений на 3 группы (6, 9, 10 чел.): {answer}")

# Задача 12. Семизначные числа из 4, 5, 6 с заданными повторами
'''
Есть семизначное число, в котором есть заданные определённые цифры (причем сами значения этих цифр не важны в рамках этой задачи). Соответственно, семь - это количество позиций нахождения каждого из них.
В таком случае, конечное количество перестановок этих цифр = 7! = 5040. Однако, это количество подразумевает уникальность каждого значения в перестановке.
Поскольку в наборе цифр есть повторяющиеся значения, то для каждого их размещения нет необходимости в уникальных позициях и мы можем убрать перестановки с повторными
цифрами, а именно сократить 7! в (2!*2!*3!)=24 раза.
'''
answer = int(factorial(7) / factorial(2) / factorial(2) / factorial(3))
print(f"12. Факториал с делением, количество перестановок с повторяющимися значениями: {answer}")
```

```python
# Классическая вероятностная модель. Геометрическая вероятность

# Задание 1.
answer = round((5/9) * (4/8) * (3/7) * 100, 2)
print(f"1. Вероятность достать 3 апельсина случайным выбором без возврата: {answer} %")

# Задание 2.
'''
Похожая задача про то, с какой вероятностью в случайной группе людей будет как минимум 2 человека, у которых д/р в один день (спойлер: при 25 чел - 56%, а уже при 60 чел - 99,5%)
Первый студент может выбрать любое значение от 1 до 10. Второй студент в 9 из 10 случаев будет несовпадать. Третий - в оставшихся 8 из 10 с двумя другими.
'''
answer = round((1 - ((10*9*8) / 10**3)) * 100, 2)
print(f"2. Вероятность совпадения числа: {answer} %")

# Задание 3.
'''
В данной задаче важное условие, что это число. Поэтому, оно не может начинаться на 0 и на первой позиции может быть 9 варинатов цифр (от 1 до 9).
Далее, нам нужно найти количество сочетаний 8-значного числа, где наше какое-то повторяющееся число будет будет встречаться в 4 из 8 позиций.
После добавляем условие уникальности оставшихся 4 цифр. Поскольку, мы уже выбрали какую-то из 10 цифр, то на выбор останется 9 вариантов. В конце
нам нужно учесть количество перестановок этих разных 4 цифр. Общее число вариантов расположения цифр в 8-значном числе 9 * 10**7 = 90 000 000.
'''
answer = (9 * comb(8, 4) * comb(9, 4) * factorial(4)) / (9 * 10**7) * 100
print(f"3. Вероятность того, что в 8-значном числе ровно 4 цифры совпадают, а остальные различны: {answer} %")

# Задание 4.
'''
Найдём вероятности чтобы все фирмы получили хотя бы 1 клиента и вычтем эту вероятность из 100%. Всего варантов выбрать 6 клиентам 5 фирм будет 5**6 = 15625.
Далее последовательно мы будем добавлять и исключать включения, пока не дойдём до варианта, где 6 клиентов идут в одну фирму из 5.
'''
answer = 1 - ((5**6 - comb(5,1)*4**6 + comb(5,2)*3**6 - comb(5,3)*2**6 + comb(5,4)*1**6) / 5**6)
print(f"4. Вероятность, что хотя бы 1 фирма останется без клиента: {answer * 100} %")
```
